### One-Hot Encoding

```python
data_new = pd.get_dummies(data_new, drop_first = True)
```

```python
#Dependent and independent features
X = data_new.iloc[:, 1:]
y = data_new.iloc[:, 0]
```

```python
X.head()
```

```python
y.head()
```

```python
#Feature importance
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X, y)
```

```python
#Finding features with greatest importance
print(model.feature_importances_)
```

```python
#Plotting a graph of feature importances for better visualization
ft_importances = pd.Series(model.feature_importances_, index = X.columns)
ft_importances.nlargest(5).plot(kind = 'barh')
plt.show()
```

```python
#Performing a train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

```python
#Hyperparameters
import numpy as np
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
print(n_estimators)
```

```python
#Randomized search CV

#Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
#Number of features to consider at every split
max_features = ['auto', 'sqrt']
#Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
#max_depth.append(None)
#Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
#Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
from sklearn.model_selection import RandomizedSearchCV #Helps find out the best parameters to use
```

```python
#Create the random grid
random_grid = {'n_estimators': n_estimators,
              'max_features': max_features,
              'max_depth': max_depth,
              'min_samples_split': min_samples_split,
              'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
#Use te random grid to search for best hyperparameters
#First create the base model to tune

rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf,
                              param_distributions = random_grid,
                              scoring = 'neg_mean_squared_error',
                              n_iter = 10,
                              cv = 5,
                              verbose = 2,
                              random_state = 42,
                              n_jobs = 1)
```

```python
rf_random.fit(X_train, y_train)
```

```python
#After the model, we now do the prediction
predictions = rf_random.predict(X_test)
```

```python
predictions
```

```python
#Compare using the distplot
sns.distplot(y_test - predictions);
```

```python
#Plotting a scatter graph
plt.scatter(y_test, predictions); #Our prediction is good as there is a linear relationship
```

```python
#Since the model is created, we store the file in form of a pickle file
import pickle

#Open a file to store the data
file = open('random_forest_regression_model.pkl', 'wb') #A pickle file is a serialized file which we'll us for deployment

#Dump information into that file
pickle.dump(rf_random, file)
```
